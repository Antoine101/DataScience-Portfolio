## Querying the database

```python
%%sql
SELECT *
FROM sqlite_master
WHERE type='table'
```

## What are the top 5 rows of the facts table?

```python
%%sql
SELECT *
FROM facts
LIMIT 5
```

## What are the minimum and maximum of population and population growth?

```python
%%sql
SELECT MIN(population) as Minimum_population,
       MAX(population) as Maximum_population,
       MIN(population_growth) as Minimum_population_growth,
       MAX(population_growth) as Maximum_population_growth
FROM facts
```

## What is the country with population of 0?

```python
%%sql
SELECT *
FROM facts
WHERE population == (SELECT MIN(population) FROM facts)
```

## Which country in the facts table corresponds to the maximum population?

```python
%%sql
SELECT *
FROM facts
WHERE population == (SELECT MAX(population) FROM facts)
```

## Re-calculation of the summary statistics without the 'World' row

```python
%%sql
SELECT MIN(population) as Minimum_population,
       MAX(population) as Maximum_population,
       MIN(population_growth) as Minimum_population_growth,
       MAX(population_growth) as Maximum_population_growth
FROM facts
WHERE name != 'World'
```

## What is the average population and area?

```python
%%sql
SELECT ROUND(AVG(population),2) as Average_population,
    ROUND(AVG(area),2) as Average_area
FROM facts
WHERE name != 'World'
```

## What countries are the most densely populated?

```python
%%sql
SELECT *
FROM facts
WHERE population > (SELECT AVG(population) FROM facts WHERE name != 'World')
and area < (SELECT AVG(area) FROM facts WHERE name != 'World')
```

## What country has the most people?

```python
%%sql
SELECT *
FROM facts
WHERE population == (SELECT MAX(population) FROM facts WHERE name != 'World')
```

## What country has the highest growth rate?

```python
%%sql
SELECT *
FROM facts
WHERE population_growth == (SELECT MAX(population_growth) FROM facts WHERE name != 'World')
```

## Which countries have the highest ratios of water to land?

```python
%%sql
SELECT name, area_water/area_land as Water_Land_Area_Ratio
FROM facts
ORDER BY Water_Land_Area_Ratio DESC
```

## Whic countries will add the most people to their population next year?

```python
%%sql
SELECT name, population, population_growth, ROUND(population*population_growth/100,1) as Next_Year_New_Population
FROM facts
WHERE name != 'World'
ORDER BY Next_Year_New_Population DESC
```

## Which countries have a higher death rate than birth rate?

```python
%%sql
SELECT name, birth_rate, death_rate
FROM facts
WHERE death_rate > birth_rate and name != 'World'
```

## What countries have the highest population/area ratio?

```python
%%sql
SELECT name, population, area_land, population/area_land as population_area_ratio
FROM facts
WHERE name != 'World'
ORDER BY population_area_ratio DESC
```
